```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Salary_dataset.csv")
data.head()
```

```python
X = data.drop(columns=['Salary'])
y = data["Salary"].values
```

```python
def gradient_descent(X, y, alpha, epoch, lam):
    m = len(y)

    x_biased = np.c_[np.ones(m), X]
    n = x_biased.shape[1]
    theta = np.zeros(n)

    tolerance = 1e-6  # how small the improvement must be
    prev_cost = 0

    for i in range(epoch):
        h = x_biased.dot(theta)
        error = h - y
        print(np.abs(theta))
        gradient = (1/m) * x_biased.T.dot(error) + ((lam/m)*np.abs(theta))
        theta -= alpha * gradient
        if i%100 == 0:
            print(theta)
        cost = (1/(2*m)) * np.sum(error**2)

        if abs(prev_cost - cost) < tolerance:
            print(f"Converged after {i} iterations.")
            break

        prev_cost = cost

    return theta
    
def prediction(X,theta):

    x_new = np.c_[np.ones(len(X)), X]
    y_pred = x_new.dot(theta)

    return y_pred
```
