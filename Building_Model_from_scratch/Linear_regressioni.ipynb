```python
def data_prep(data,target):
    X = data.drop(columns=[target])
    y = data[target].values

    return X, y
```

# Calculating best Theta using Gradient Update

```python
def gradient_descent(X, y, L, iters):
    m = len(y)

    # X = (X - X.mean()) / X.std()

    X_bias = np.c_[np.ones(m), X]
    print(X_bias.shape)

    n = X_bias.shape[1]
    theta = np.zeros(n)


    tolerance = 1e-6  # how small the improvement must be
    prev_cost = 0

    for i in range(iters):
        h = X_bias.dot(theta)
        error = h - y
        gradient = (1/m)*X_bias.T.dot(error)
        theta -= L * gradient
        cost = (1/(2*m)) * np.sum(error**2)
        

        # if i%100 == 0:
        #     print(f"{i}th cost is: {cost:.4f}")

        if abs(prev_cost - cost) < tolerance:
            print(f"Converged after {i} iterations.")
            break

        prev_cost = cost

    return theta


def prediction(X,theta):

    x_new = np.c_[np.ones(len(X)), X]
    y_pred = x_new.dot(theta)

    return y_pred
```

# Calculating Theta Using Newton Method

```python
def newton_method(X,y):
    m = len(y)

    # X = (X - X.mean()) / X.std()

    X_bias = np.c_[np.ones(m), X]
    print(X_bias.shape)

    
    theta = np.linalg.solve(X_bias.T @ X_bias, X_bias.T @ y)
    print(theta)

newton_method(X,y)

    
```
