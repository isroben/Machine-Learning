```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('insurance_data.csv')
```

```python
def data_prep(data,target):
    X = data.drop(columns=[target])
    y = data[target].values

    return X, y
X, y = data_prep(df, "bought_insurance")
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))

def calculate_gradient(theta, X, y):
    m = y.size
    return (X.T @ (sigmoid(X @ theta) - y))/m

def gradient_descent(X, y, L=0.1, max_iter=100000, tol=1e-7):
    m = len(y)
    x = np.c_[np.ones(m), X]
    n = x.shape[1]
    theta  = np.zeros(n)
    
    for i in range(max_iter):
        grad = calculate_gradient(theta, x, y)
        theta -= L * grad

        if np.linalg.norm(grad) < tol:
            print(f"Converged in {i}th time with gradient: {grad}")
            break

    return theta

def predict_prob(X, theta):
    m = X.shape[0]
    x = np.c_[np.ones(m), X]
    return sigmoid(x@theta)

def predict(X, theta, threshold=0.5):
    return (predict_prob(X, theta) >= threshold).astype(int)

```

```python
theta_hat  = gradient_descent(X, y, 0.01)
print(theta_hat)
```
